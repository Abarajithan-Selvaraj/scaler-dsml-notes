```python
# D36 - Q3: Given a data frame consisting of Name, employment_type, department, and p_id for employees. Complete the function del_dupl() that

# 1. Deletes the records of a given name, where name is a string, then
# 2. Deletes the duplicate p_id rows and keeps only the last entry as it contains the latest employment details of the candidate
# 3. Deletes the column p_id as the organization is moving to another fintech app, and returns this dataframe.

# There are no two people with the same p_id in the organization, if there are records with the same p_id then it should be assumed that the person has updated their profile (designation) in the organization.

def del_dupl(df,name):
    ''' df is a dataframe with columns ['Name', 'employment_type', 'department', 'p_id']
        name is a string.
        Output a dataframe is expected to be returned'''
        
    # YOUR CODE GOES HERE
    # delete the record with name
    ind = df.loc[df["Name"] == name].index
    df.drop(ind, inplace = True)
    # drop the column 'p_id'
    df.drop(columns=["p_id"], inplace=True)
    # drop the duplicates, preserving the last one with same name
    df.drop_duplicates(subset=["Name"], keep="last", inplace=True)
    
    return df
```

```python
# D37 - Q6: Given a data frame as shown below, return the series of male names in the age group of 23-30 (inclusive of both 23 and 30).

def specific_customers(df):
    ''' df is a dataframe with columns ['name', 'profession', 'gender', 'age', 'review', 'rating']
        Output -> A series with required names is expected to be returned
    '''
    # YOUR CODE GOES HERE    
    # Select male professionals
    df_new = df.loc[df["gender"] == "male"]
    
    # Filter for age group 23-30
    df_new = df_new.loc[(df["age"] >= 23) & (df["age"] <= 30)]
    
    return df_new['name']
```

```python
# D37 - Q7: Given a dataframe containing records of student and their marks, and cutoff marks k.
# Return the name of students who scored marks >= k

def filter(df, k):
    '''
    INPUT: 
        df -> dataframe
        k -> cutoff (int)    
    OUTPUT:
        result -> dataframe
    '''
    
    ### STEP 1 : Create mask
    mask = df["marks"] >= k
    ### STEP 2: Use mask to filter rows
    result = df.loc[mask]["name"]
    
    return result
```

```python
# D37 - Q8: Given a dataframe containing employee income information, Calculate & return the percentage of employees in medium income bracket.

def calculate(df):
    '''
    INPUT: df -> pandas dataframe
    OUTPUT: result -> float
    '''
    
    ### STEP 1 : filter employees with medium income
    medium_income = df.loc[df["income"] == "medium"]
    
    ### STEP 2: calculate count of medium income employees
    medium_income_count = medium_income.shape[0]
    
    ### STEP 3: calculate percentage of medium income employees
    result = medium_income_count * 100 / df.shape[0]

    return result
```

```python
# D37 - Q9: Given a dataframe consisting of income details, Return the name of person having the highest income.

def max_income(df):
    '''
    INPUT: df -> dataframe
    OUTPUT: result -> String
    '''
    
    ### STEP 1: Get the row with maximum income
    max_income = df.loc[df["income"] == df["income"].max()]
    
    ### STEP 2: On above result, Filter the name column and extract string value from it
    result = df.loc[max_income.index[0], "name"]
    
    return result
```

```python
# D37 - Q10: Given data frame consists of the following information about different countries

# Population,
# The year it was measured, and
# The continent it belongs to
# Complete the function population_df() to return a data frame consisting of records where the population is greater than 10 million.

# The data frame should be sorted in ascending order, first by year and then by population columns.

def population_df(df):
    '''df is a dataframe with columns ['Country', 'Year', 'Population', 'Continent']
       Output -> A dataframe is expected to be returned.'''
       
    # Filter the dataframe for records greater than 10 million
    df_filtered = df.loc[df["Population"] > 10000000]
    
    # Sort the filtered dataframe first by year, then by population
    df_sorted = df_filtered.sort_values(["Year", "Population"], ascending=[True, True])
    
    # Don't modify this line of code    
    return df_sorted.reset_index(drop = True)
```

```python
# D39 - Q5: Given two dataframes df1 (containing details of customer) and df2(having details of orders),
# Merge the dataframe such that resultant dataframe contains records for customer ids which are common in both dataframe.

def merge(df1, df2):
    '''
    INPUT: df1, df2 -> dataframe
    OUTPUT: result -> dataframe
    '''
    
    result = pd.merge(df1, df2)
    return result
```

```python
# D39 - Q6: Given two dataframes customer (containing details of customer) and orders (having details of orders).
# Also, name is given which is a customer's name.
# Perform the following operation:
#     Merge the dataframe such that the resultant dataframe should contain records of all customer ids present in customer dataframe.
#     Calculate the total order amount for given customer name
# Return the merged dataframe and the sum amount.

def get_amount(customer, orders, name):
    '''
    INPUT: 
        customer, orders -> dataframe
        name -> string  
    OUTPUT: 
        (merged_df, sum_amount) -> tuple of (dataframe, float)
    '''
    
    ## STEP 1: merge customer and orders dataframe
    merged_df = pd.merge(customer, orders, how="left")
    ## STEP 2: calculate sum of amount for given name
    sum_amount = merged_df[merged_df["name"] == name]["amount"].sum()

    return merged_df, sum_amount
```

```python
# D39 - Q9: Given a data frame containing area and population data, calculate the Population density for each state.
# The function should return Series of population densities sorted in ascending order.
# Note: Population density is defined as population per unit area.

def calc_density(df): 
    '''
    INPUT: df -> dataframe
    OUTPUT: result -> A series
    '''
    
    ## STEP 1: calculate population density
    pop_den = df.apply(lambda x: x["population"]/x["area"], axis=1)
    ## STEP 2: sort the values
    result = pop_den.sort_values()
    
    return result
```

```python
# D40 - Q4: Given a dataset about salesperson and customer orders, the task is to group the data as per salesman_id and customer_id, and return a new data frame with the ord_no count for each salesman_id and customer_id pair.

def group_the_data(df):
    '''
    input:
    df -> the dataframe provided to the function
    output:
    result -> the grouped by data as per required in the question
    '''
    result = None    
    # Your Code Starts here
    result = df.groupby(by=["salesman_id", "customer_id"])["salesman_id", "customer_id", "ord_no"].aggregate(ord_no=("ord_no", "count")).reset_index()
    # Your Code ends here
    return result
```

```python
# D40 - Q5: Given sales data of Amazon, our task is to filter out user data from this dataset where the user has bought a total minimum of "x" quantity of items, and also the average price of items bought by this user should be atleast "y"$. Values of x and y will be provided to the function.

def filter_the_customers(df, x, y):
    '''
    Input:
    df -> The input sales data
    x -> The minimum quantity required
    y -> The average price required
    Output:
    Return the filtered dataframe
    '''
    # Your Code Starts here
    df["Price"] = df["Price Each"] * df["Quantity Ordered"]
    
    df1 = df.groupby(by=["Name"]).aggregate(qty=("Quantity Ordered", "sum"), total_price=("Price", "sum")).reset_index()
    df1["avg_price"] = df1["total_price"] / df1["qty"]
    df1 = df1.loc[(df1["qty"] >= x) & (df1["avg_price"] >= y)]
    
    # df = df.merge(df1, how="inner", on="Name")
    df = df.loc[lambda x: x["Name"].isin(df1["Name"])]
    
    return df.loc[:, ["Name","Purchase Address"]]
```

```python
# D41 - Q3: Given the data frame df as input, do the following steps for preprocessing:
# 1. Remove the row if all the columns have missing values.
# 2. Replace the missing values of "Roll_ID" column with 0 and "Name" column with "Anonymous"
# 3. Replace the missing values in "Marks" column with the median value of the column
# 4. Change the numerical columns (Roll_ID and Marks) to int datatype in the output

def preprocess(df):
  """
    Takes df as input, do the preprocessing as asked.
    Return the preprocessed dataframe
  """
  # Drop the rows with all null values
  df.dropna(how="all", inplace=True)
  # Replace the missing values of "Roll_ID" column with 0 and "Name" column with "Anonymous"
  alias = {"Roll_ID": 0, "Name": "Anonymous"}
  df.fillna(value=alias, inplace=True)
  # Replace the missing values in "Marks" column with the median value of the column
  median = df["Marks"].median()
  df["Marks"].fillna(median, inplace=True)
  # Change the numerical columns (Roll_ID and Marks) to int datatype in the output
  df["Roll_ID"] = df["Roll_ID"].astype(int)
  df["Marks"] = df["Marks"].astype(int)
  
  return df

df = pd.read_csv("./img-3.csv")
preprocess(df)
```

```python
# D42 -Q1: Given a dataframe consisting of organization's access id, return the count of invalid access ids.
# The valid access ids should contain 'ORG' in the beginning.

# For example: 
# Valid: 'ORG1234' / 'ORG4200'
# Invalid: '1234' / 'OR1435'

def count_ids(df):
    '''
    INPUT: df -> dataframe
    OUTPUT: result -> integer
    '''
    
    ## STEP1 : Get the valid ids mask    
    valid_id_mask = df["access_id"].str.startswith("ORG")
    
    ## STEP 2: Get invalid ids    
    invalid_ids = df.loc[valid_id_mask == False]
    
    ## STEP 3: Get count of invalid ids    
    result = invalid_ids.shape[0]
    
    return result 
```

```python
# D42 - Q2: Given a dataframe of student's username for batch of 2022,

# return the count of username which doesn't follow the below given guidelines:

#     The username should start with 'user_', and
#     The username should end with '_2022'


def validate(df):
    '''
    INPUT: df -> dataframe    
    OUTPUT: result -> int
    '''
    
    ## STEP 1: Get the mask for vaild username    
    valid_username_mask = df["username"].str.startswith("user_") & df["username"].str.endswith("_2022")
    
    ## STEP 2: Find the invalid usernames    
    invalid_usernames = df.loc[valid_username_mask == False]
    
    ## STEP 3: Get count of invalid usernames    
    result = invalid_usernames.shape[0]
    
    return result 
```

```python
# D42 - Q3: Given a location dataframe, with a single column "City/tState", return the preprocessed dataframe with names of City and State separated.

def location_divide(df):
    '''
    input:
    df -> the dataframe provided to the function    
    output:
    df_new -> the new preprocessed dataframe
    '''
    
    ### STEP 1: Split the values into 2 columns    
    df_new = df["City\tState"].str.split("\t", expand=True)
    
    ### STEP 2: Split the column name into 2    
    df_new.columns = ["City", "State"]
    
    return df_new
```

```python
# D43 - Q5: You're given a record of students with their registration dates and registration ids. You're also given the marks scored by each of the students in Physics, Chemistry, and Mathematics. You need to find out the month that observed the maximum registrations. Then return the average marks scored in each subject. In case of a tie, select the month that comes first (lexicographically).

df = pd.DataFrame({'Date':["2015-12-06", "2011-12-27", "2015-09-07", "2012-12-21", "2020-02-13", "2015-06-09"], 'RID':[498, 721, 375, 464, 813, 853], 'Phy':[22, 45, 1, 65, 22, 17], 'Chem':[52, 56, 32, 50, 24, 61], 'Math':[63, 37, 68, 62, 43 ,42]}) 

df["Date"] = df["Date"].astype("datetime64")
df["Month"] = df["Date"].dt.strftime("%b").str.upper()
dff = df.groupby(["Month"]).agg(count_regs=("Month", "count"), mean_chem=("Chem", "mean"), mean_phy=("Phy", "mean"), mean_math=("Math", "mean")).reset_index().sort_values(by="count_regs", ascending=False)

dff = dff.round(2)

dff.iloc[0]
```

```python
# D42 - Q6: As an educational institute, you need to keep a track of all the registered students. Here you're given the registration IDs and the corresponding dates of a batch of students. You need to return a DataFrame containing the columns as follows:

# RID: Id of the student
# RDate: Date of joining of the student
# RMonth: Month of joining of the student
# RYear: Year of joining of the student
# RDay: Day of joining of the student

def solve(reg_id, reg_dates):
    res_df = pd.DataFrame(zip(reg_id, reg_dates), columns=["RID", "RDate"])
    """
    input:
    reg_id -> id of the input dataframe
    reg_dates -> dates of the input dataframe
    res_df -> the input dataframe 
    
    output:
    return the resultant dataframe after performing the required operation
    """
    # YOUR CODE GOES HERE
    res_df["RDate"] = res_df["RDate"].astype("datetime64")
    res_df["RMonth"] = res_df["RDate"].dt.month
    res_df["RYear"] = res_df["RDate"].dt.year
    res_df["RDay"] = res_df["RDate"].dt.day
    
    return res_df
```
